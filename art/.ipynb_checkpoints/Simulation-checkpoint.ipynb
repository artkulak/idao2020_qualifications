```python
import plyades
import numpy as np
import astropy
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
vec = [9802.903469909521, -4033.0140760242025, -2068.3038998721813,
       -2.596979067630373, -4.475850167940195, -3.581083217188638]
```

```python
x = vec[0]
y = vec[1]
z = vec[2]
vx = vec[3]
vy = vec[4]
vz = vec[5]
iss_r = np.array([
x,
y,
z,
]) * astropy.units.km
iss_v = np.array([
vx,
vy,
vz,
]) * astropy.units.km/astropy.units.s
iss_t = astropy.time.Time('2014-01-01T00:00:00.000')
frame = 'ECI'
body = plyades.bodies.EARTH
```

```python
@property
def elements(self):
    return kepler.elements(self.body.mu, self.r, self.v)
```

```python
iss = plyades.State(iss_r, iss_v, iss_t, frame, body)
```

```python
@iss.gravity
def newton_j2(f, t, y, params):
    r = np.sqrt(np.square(y[:3]).sum())
    mu = params['body'].mu.value
    j2 = params['body'].j2
    r_m = params['body'].mean_radius.value
    rx, ry, rz = y[:3]
    f[:3] += y[3:]
    pj = -3/2*mu*j2*r_m**2/r**5
    f[3] += -mu*rx/r**3 + pj*rx*(1-5*rz**2/r**2)
    f[4] += -mu*ry/r**3 + pj*ry*(1-5*rz**2/r**2)
    f[5] += -mu*rz/r**3 + pj*rz*(3-5*rz**2/r**2)
```

```python
j2_orbit = iss.propagate(iss.period*600, max_step = 1000, interpolate=1000)
```

```python
fig1 = go.Scatter3d(x=pd.DataFrame(j2_orbit.rx).values.reshape(-1), 
                    y=pd.DataFrame(j2_orbit.ry).values.reshape(-1), 
                    z=pd.DataFrame(j2_orbit.rz).values.reshape(-1), 
                    mode='markers', 
                    marker=dict(
                        #color='rgb(127, 127, 127)',
                        color = np.arange(pd.DataFrame(j2_orbit.rx).values.shape[0]),
                        colorscale = 'Viridis',
                        showscale = True,
                        size=8, 
                        symbol='circle', 
                        line=dict(
                            #color='rgb(204, 204, 204)',
                            width=0.5),
                        opacity=0.9),
                   name = 'Real')
```

```python
fig = go.Figure(data = [fig1])
fig.show()
```
